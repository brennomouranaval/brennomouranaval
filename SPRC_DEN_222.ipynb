```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import sqlite3 as sql
import os
import datetime as datetime
```

```python
df_sprc_222 = pd.read_excel('sprc_222_17022025.xlsx',header=0)
df_sprc_222
```

```python
# Remove duplicadas
#df_sprc_222 = df_sprc_222.drop_duplicates(subset='NUM DO PS')
#df_sprc_222
```

```python
nao_prontos = df_sprc_222[df_sprc_222['PERC. CONC.'] != 1.0].count()
display(nao_prontos['PERC. CONC.'])

prontos = df_sprc_222[df_sprc_222['PERC. CONC.'] == 1.0].count()
display(prontos['PERC. CONC.'])

totais = df_sprc_222.count()
display(totais['PERC. CONC.'])
```

```python
df_nao_prontos = df_sprc_222[df_sprc_222['PERC. CONC.'] != 1.0]
df_nao_prontos
```

```python
# Gráfico de pizza utilizando plotly
labels = ['Prontos', 'Não Prontos']
values = [prontos['PERC. CONC.'], nao_prontos['PERC. CONC.']]
fig = go.Figure(data=[go.Pie(labels=labels,
                             values=values,
                             hole=.3,
                             pull=[0, 0.2],
                             textinfo='label+percent',
                             )])

fig.update_layout(title='Percentual de conclusão de OS do DEN-222',
                  title_x=0.5,
                  plot_bgcolor='rgba(0,0,0,0)',
                  paper_bgcolor='rgba(0,0,0,0)',
                  template='plotly_dark')

fig.show()
```

```python
df_nao_prontos_nao_duplicados = df_nao_prontos.drop_duplicates(subset='NUM DO PS')
df_nao_prontos_nao_duplicados
```

```python
atividades_executor = df_nao_prontos_nao_duplicados['EXECUTOR'].value_counts()
atividades_executor
```

```python
atividades_executor_sorted = atividades_executor.sort_values(ascending=True)

fig = go.Figure(data=[go.Bar(
    x=atividades_executor_sorted.values,
    y=atividades_executor_sorted.index,
    orientation='h'
)])
fig.update_layout(title='Atividades não concluídas por executor',
                  title_x=0.5,
                  xaxis_title='Quantidade de atividades',
                  yaxis_title='Executor',
                  plot_bgcolor='rgba(0,0,0,0)',
                  paper_bgcolor='rgba(0,0,0,0)',
                  template='plotly_dark')

fig.show()


atividades_executor_sorted = atividades_executor.sort_values(ascending=True)

# Radar chart
fig = go.Figure(data=go.Scatterpolar(
    r=atividades_executor_sorted.values,
    theta=atividades_executor_sorted.index,
    fill='toself',  # Preenche a área do gráfico
    name='Atividades não concluídas'
))

fig.update_layout(
    title='Atividades não concluídas por executor',
    title_x=0.5,
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, max(atividades_executor_sorted.values) * 1.1]  # Ajuste o valor do range conforme necessário
        )
    ),
    showlegend=False,
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    template='plotly_dark'
)

fig.show()
```

```python
tipo_documento = df_nao_prontos_nao_duplicados['DOCUMENTO A SER GERADO'].value_counts()
tipo_documento
```

```python
tipo_documento_sorted = tipo_documento.sort_values(ascending=True)

# Radar chart
fig = go.Figure(data=go.Scatterpolar(
    r=tipo_documento_sorted.values,
    theta=tipo_documento_sorted.index,
    fill='toself',  # Preenche a área do gráfico
    name='Atividades não concluídas'
))

fig.update_layout(
    title='Atividades não concluídas por executor',
    title_x=0.5,
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, max(tipo_documento_sorted.values) * 1.1]  # Ajuste o valor do range conforme necessário
        )
    ),
    showlegend=False,
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    template='plotly_dark'
)

fig.show()
```

```python
data_hoje = datetime.datetime.now()

df_nao_prontos_nao_duplicados['Dias Restantes'] = (df_nao_prontos_nao_duplicados['DATA LIMITE'] - data_hoje).dt.days
df_nao_prontos_nao_duplicados
```

```python
# Ordenar df_nao_prontos por 'Dias Restantes' em ordem decrescente
df_nao_prontos_nao_duplicados = df_nao_prontos_nao_duplicados.sort_values(by=['Dias Restantes'], ascending=False)

fig = go.Figure(data=[go.Bar(
       x=df_nao_prontos_nao_duplicados['Dias Restantes'][df_nao_prontos_nao_duplicados['Dias Restantes'] > 0],
       y=df_nao_prontos_nao_duplicados['NUM DO PS'][df_nao_prontos_nao_duplicados['Dias Restantes'] > 0],
       orientation='h',
       text=df_nao_prontos_nao_duplicados['Dias Restantes'][df_nao_prontos_nao_duplicados['Dias Restantes'] > 0],
       marker=dict(
           color=df_nao_prontos_nao_duplicados['Dias Restantes'][df_nao_prontos_nao_duplicados['Dias Restantes'] > 0],
           colorscale='RdBu',  # Specify the colorscale
           colorbar=dict(title='Dias restantes')  # Specify the colorbar title
       )
   )])

fig.update_layout(
       title='Dias restantes para conclusão das atividades',
       title_x=0.5,
       plot_bgcolor='rgba(0,0,0,0)',
       paper_bgcolor='rgba(0,0,0,0)',
       height=600,
       width=1000,
       template='plotly_dark'
   )

fig.show()
```
